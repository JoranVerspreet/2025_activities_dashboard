## Print nbr of unique values

```python
# Display column names, their data types, and the number of unique values:
print("Column Names and Data Types are:")
for column in df.columns:
 print(f"{column} datatype is: {df[column].dtype}")
 print(f'The number of unique values of {column} is: {len(df[column].unique())}')
```
